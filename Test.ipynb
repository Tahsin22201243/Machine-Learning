1. Create an array with values from 0 to 9 and extract all odd numbers.

```python
import numpy as np

# Create array
arr = np.arange(0, 10)
print("Original Array:", arr)
print("Shape:", arr.shape)

# Extract odd numbers
odd_numbers = arr[arr % 2 != 0]
print("Odd Numbers:", odd_numbers)
print("Shape:", odd_numbers.shape)
```

2. Extract the first two rows and first two columns from a 4×4 NumPy array.


```python
# Create 4x4 array
arr_4x4 = np.arange(1, 17).reshape(4, 4)
print("Original Array:\n", arr_4x4)
print("Shape:", arr_4x4.shape)

# Extract first 2 rows and first 2 columns
sub_array = arr_4x4[:2, :2]
print("Extracted Sub-array:\n", sub_array)
print("Shape:", sub_array.shape)
```

3. Filter DataFrame rows where "Age" > 25.

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [22, 28, 35, 24],
    "Salary": [30000, 40000, 50000, 35000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Filter rows
filtered_df = df[df["Age"] > 25]
print("\nFiltered DataFrame (Age > 25):\n", filtered_df)
print("Shape:", filtered_df.shape)
```

4. Select only "Name" and "Salary" columns from existing DataFrame.

```python
selected_columns = df[["Name", "Salary"]]
print("Selected Columns:\n", selected_columns)
print("Shape:", selected_columns.shape)
```

5. Find and print unique values in a specific column.

```python
unique_values = df["Age"].unique()
print("Unique Age Values:", unique_values)
print("Shape:", unique_values.shape)
```
